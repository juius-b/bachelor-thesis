```python
import os
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Define the directory path where your images are stored
directory = "/mnt/jbrockma/bachelor-thesis-images/chest"
```

```python
# Get a list of all the image file names in the directory
image_files = os.listdir(directory)
```

```python
# Choose a random subset of images from the list
num_images_to_display = 10  # Change this value to the number of images you want to display
selected_images = random.sample(image_files, num_images_to_display)
plt.figure(figsize=(20, 10))  # Adjust the figure size as per your preference
for i, image_name in enumerate(selected_images):
    image_path = os.path.join(directory, image_name)
    img = mpimg.imread(image_path)

    # Get the image width and height
    width, height = img.shape[1], img.shape[0]

    plt.subplot(2, 5, i+1)  # Adjust the subplot configuration as per your preference
    plt.imshow(img)
    plt.title(f"{width == height}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```
