```python
import pandas as pd
```

```python
df = pd.read_csv("/mnt/jbrockma/CXR8/Data_Entry_2017_v2020.csv")
df.head()
```

```python
df = df.rename(columns={
    "Image Index": "file_name",
    "Finding Labels": "findings"
})
df.head()
```

```python
df = df[["file_name", "findings"]]
df.head()
```

```python
finding_dummies = df["findings"].str.get_dummies("|")
finding_dummies.head()
```

```python
finding_dummies = finding_dummies.rename(columns=str.lower)
finding_dummies.head()
```

```python
finding_dummies = finding_dummies.drop("no finding", axis=1)
finding_dummies.head()
```

```python
df = pd.concat([df["file_name"], finding_dummies], axis=1)
df.head()
```
